```python
import re
```

```python
f = open("./data/good.txt", "r")
if f.mode == 'r':
    contents = f.read()

abono_regex = re.compile("(Abono DE PERMANENCIA (ao|equiva)[\s\S]*?\.\n)", re.IGNORECASE)
substituicao_regex = re.compile("(DESIGNAR.*?para\ssubstituir[\s\S]*?\.\n)", re.IGNORECASE)
retificar_regex = re.compile("(RETIFICAR, [\s\S]*?\.\n)", re.IGNORECASE)


for match in re.finditer(substituicao_regex, contents):
    print((match.group()))
```
