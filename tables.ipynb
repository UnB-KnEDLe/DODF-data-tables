```python
from IPython.display import HTML
import random

def hide_toggle(for_next=False):
    this_cell = """$('div.cell.code_cell.rendered.selected')"""
    next_cell = this_cell + '.next()'

    toggle_text = 'Toggle show/hide'  # text shown on toggle link
    target_cell = this_cell  # target cell to control with toggle
    js_hide_current = ''  # bit of JS to permanently hide code in current cell (only when toggling next cell)

    if for_next:
        target_cell = next_cell
        toggle_text += ' next cell'
        js_hide_current = this_cell + '.find("div.input").hide();'

    js_f_name = 'code_toggle_{}'.format(str(random.randint(1,2**64)))

    html = """
        <script>
            function {f_name}() {{
                {cell_selector}.find('div.input').toggle();
            }}

            {js_hide_current}
        </script>

        <a href="javascript:{f_name}()">{toggle_text}</a>
    """.format(
        f_name=js_f_name,
        cell_selector=target_cell,
        js_hide_current=js_hide_current, 
        toggle_text=toggle_text
    )

    return HTML(html)

hide_toggle()
```

## Tabela com repetições

```python
tables_array = [aposentadoria, exoneracao, nomeacao_comissionados, atos_sem_efeito, retificacoes, reversao,
                atos_sem_efeito_aposentadoria, abono_de_permanencia, retificacoes_aposentadoria,
                exoneracao_efetivos, substituicao_de_funcao, cessoes, nomeacao_efetivos, reposicionamento]
table_with_repetitions = pd.concat(tables_array, ignore_index=True)
```

## Tabela sem repetições

```python
# Table with duplicates dropped
table_without_duplicates = table_with_repetitions.drop_duplicates(subset="Atributo", 
                                                                  keep="first", 
                                                                  ignore_index=True)
```

## Gráfico Obrigatoriedade

```python
# Tuple with key and value (Obrigatoriedade, Repetições)
madatory_repetitions = table_with_repetitions["Obrigatoriedade"].value_counts()
madatory_repetitions
```

```python
labels = ["Obrigatório", "Não Obrigatório"]
sizes = [madatory_repetitions[0], madatory_repetitions[1]]
colors = ['#86BA90', '#F5F3BB']
fig1, ax1 = plt.subplots(figsize=(8,8))
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90, colors=colors)
ax1.axis("equal")
ax1.set_title("Taxa de obrigatoriedade dos atributos")
ax1.legend(labels, title='Obrigatoriedade', 
           loc='center left', 
           bbox_to_anchor=(1.3, 0, 0.5, 1))
plt.show()
```

## Gráfico Repetição de Atributos

```python
# Tuple with key and value (Atributo, Repetições)
repeted_attributes = table_with_repetitions["Atributo"].value_counts()
repeted_attributes
```

```python
figure(num=None, figsize=(21, 9), dpi=80, facecolor='w', edgecolor='k')

labels = []
sizes = []
count = 0
keys = repeted_attributes.keys()


for i in range(0, len(list(repeted_attributes))):
    if repeted_attributes[i] > 1:
        sizes.append(repeted_attributes[i])
        count += 1

for j in range(0, count):
    labels.append(keys[j])
    
    
def plt_bar():
    index = np.arange(len(labels))
    plt.barh(index, sizes)
    plt.xlabel('N° de repetições', fontsize=8)
    plt.ylabel('Atributos', fontsize=8)
    plt.yticks(index, labels, fontsize=8, rotation=0)
    plt.title('Repetição de Atributos')
    plt.show()
    
plt_bar()
```

## Gráfico porcentagem de repetição de atributos

```python
# Quantidade de atributos diferentes que existem
total_attributes = len(repeted_attributes)

# Quantidade de atributos que se repetem
total_repeted = len(sizes)

print(f'Total de atributos diferentes: {total_attributes}')
print(f'Total de atributos que aparecem mais de uma vez:  {total_repeted}')
```

```python
labels = ["Se repetem", "Não se repetem"]
sizes = [total_repeted, total_attributes - total_repeted]
colors = ['#60606F', '#A5A475']
fig1, ax1 = plt.subplots(figsize=(8,8))
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90, colors=colors)
ax1.axis("equal")
ax1.set_title("Taxa de obrigatoriedade dos atributos")
ax1.legend(labels, title='Obrigatoriedade', 
           loc='center left', 
           bbox_to_anchor=(1.3, 0, 0.5, 1))
plt.show()
```

## Gráfico porcentagem que possui padrão

```python
# Tuple with key and value (Padrão, Repetições)
pattern = table_with_repetitions["Padrão"].value_counts()
pattern
```

```python
total_rows_table_with_repetitions = madatory_repetitions[0] + madatory_repetitions[1]
total_rows_table_with_repetitions
```

```python
labels = ["Com padrão", "Sem padrão"]
sizes = [total_rows_table_with_repetitions - pattern[0], pattern[0]]
colors = ['#9B9987', '#685762']
fig1, ax1 = plt.subplots(figsize=(8,8))
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90, colors=colors)
ax1.axis("equal")
ax1.set_title("Taxa de obrigatoriedade dos atributos")
ax1.legend(labels, title='Obrigatoriedade', 
           loc='center left', 
           bbox_to_anchor=(1.3, 0, 0.5, 1))
plt.show()
```
